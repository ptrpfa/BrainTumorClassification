# 1. Import necessary libraries and configuration

```python
import numpy as np 
import matplotlib.pyplot as plt 
from tqdm import tqdm
import cv2
import os
import seaborn as sns   
import tensorflow as tf
import pickle

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

from keras.models import Sequential, Model
from keras.layers import Dense, Flatten, Conv2D, Reshape, Input, Conv2DTranspose
from keras.layers import Activation, LeakyReLU, BatchNormalization, Dropout, Resizing
from keras.losses import BinaryCrossentropy
from tensorflow.keras.applications import VGG16

import warnings
warnings.filterwarnings('ignore')

try:
    from tensorflow.keras.optimizers import Adam
except:
    from keras.optimizers import Adam

# Ensure TensorFlow uses the GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
print(gpus)
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)
```

# 2. Global Variables and configurations

```python
NOISE_DIM = 100  
BATCH_SIZE = 32
STEPS_PER_EPOCH = 3750
EPOCHS = 10
SEED = 2200692
WIDTH, HEIGHT, CHANNELS = 128, 128, 1
NUM_CLASSES = 4  # Number of classes

# OPTIMIZER = Adam(0.0002, 0.5)
generator_optimizer = Adam(lr=0.0003, beta_1=0.5)
discriminator_optimizer = Adam(lr=0.0001, beta_1=0.5)
```

```python
# MAIN_DIR = "dataset_1/"
MAIN_DIR = "../dataset/new_dataset_1/"
```

# 3. Data Augmentation

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

# Create an ImageDataGenerator for data augmentation and normalization
datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    validation_split=0.2  # Reserve 20% of data for validation
)

# Create training and validation data generators
train_generator = datagen.flow_from_directory(
    MAIN_DIR,
    target_size=(WIDTH, HEIGHT),
    batch_size=BATCH_SIZE,
    color_mode='grayscale',
    class_mode='sparse',
    subset='training',
    seed=SEED
)

validation_generator = datagen.flow_from_directory(
    MAIN_DIR,
    target_size=(WIDTH, HEIGHT),
    batch_size=BATCH_SIZE,
    color_mode='grayscale',
    class_mode='sparse',
    subset='validation',
    seed=SEED
)
```

# 4. Buiding the GAN Model

## Generator Architecture

```python
from keras.layers import Embedding, multiply
def build_generator():

    """
        Generator model "generates" images using random noise and class labels.
        The random noise AKA Latent Vector is sampled from a Normal Distribution.
        Using Transposed Convolution, the latent vector is transformed to produce an image.
        We use Conv2DTranspose layers to generate the image features.

        Input: Random Noise / Latent Vector and class label
        Output: Image
    """
    
    noise = Input(shape=(NOISE_DIM,))
    label = Input(shape=(1,), dtype='int32')
    
    label_embedding = Flatten()(Embedding(NUM_CLASSES, NOISE_DIM)(label))
    
    model_input = multiply([noise, label_embedding])
    
    x = Dense(16 * 16 * 256)(model_input)
    x = Reshape((16, 16, 256))(x)

    x = Conv2DTranspose(128, (4, 4), strides=(2,2), padding='same')(x)
    # x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    x = Conv2DTranspose(128, (4, 4), strides=(2,2), padding='same')(x)
    # x = BatchNormalization()(x)
    x = LeakyReLU(alpha=0.2)(x)

    x = Conv2DTranspose(128, (4, 4), strides=(2,2), padding='same')(x)
    x = LeakyReLU(alpha=0.2)(x)
    
    # x = Conv2DTranspose(64, (3, 3), strides=1, padding='same')(x)
    # x = LeakyReLU(alpha=0.2)(x)

    # x = Conv2DTranspose(32, (4, 4), strides=1, padding='same')(x)
    # x = LeakyReLU(alpha=0.2)(x)

    # x = Conv2DTranspose(16, (3, 3), strides=1, padding='same')(x)
    # x = LeakyReLU(alpha=0.2)(x)

    img = Conv2D(CHANNELS, (7, 7), padding='same', activation='tanh')(x)

    model = Model([noise, label], img, name="generator")
    model.summary()
    return model
```

## Discriminator Architecture

```python
from keras.layers import Embedding, multiply
def build_discriminator():
    
    """
        Discriminator model classifies the generated images as fake or real and assigns class labels.
        It uses Convolutional layers to learn features from the images and output the validity and class labels.

        Input: Generated / Real Image and class label
        Output: Validity of Image (Fake or Real) and class label
    """
    
    img = Input(shape=(WIDTH, HEIGHT, CHANNELS))
    label = Input(shape=(1,), dtype='int32')
    
    label_embedding = Flatten()(Embedding(NUM_CLASSES, np.prod((WIDTH, HEIGHT, CHANNELS)))(label))
    label_embedding = Reshape((WIDTH, HEIGHT, CHANNELS))(label_embedding)
    
    concatenated_input = multiply([img, label_embedding])
    
    x = Conv2D(64, (3, 3), strides=2, padding='same')(concatenated_input)
    x = LeakyReLU(alpha=0.2)(x)
    
    x = Conv2D(128, (3, 3), strides=2, padding='same')(x)
    x = LeakyReLU(alpha=0.2)(x)
    
    x = Conv2D(128, (3, 3), strides=2, padding='same')(x)
    x = LeakyReLU(alpha=0.2)(x)
    
    x = Conv2D(256, (3, 3), strides=2, padding='same')(x)
    x = LeakyReLU(alpha=0.2)(x)
    
    x = Flatten()(x)
    x = Dropout(0.4)(x)
    
    validity = Dense(1, activation='sigmoid')(x)
    label_output = Dense(NUM_CLASSES, activation='softmax')(x)
    
    model = Model([img, label], [validity, label_output], name="discriminator")
    model.summary()
    model.compile(loss=['binary_crossentropy', 'sparse_categorical_crossentropy'],
                  optimizer=discriminator_optimizer)
    return model
```

## Combined GAN Model

```python
print('\n')
discriminator = build_discriminator()
print('\n')
generator = build_generator()

noise = Input(shape=(NOISE_DIM,))
label = Input(shape=(1,), dtype='int32')

# Freeze discriminator weights
discriminator.trainable = False

fake_image = generator([noise, label])
validity, target_label = discriminator([fake_image, label])

gan = Model([noise, label], [validity, target_label], name="gan_model")
gan.compile(loss=['binary_crossentropy', 'sparse_categorical_crossentropy'], optimizer=generator_optimizer)

print("The Combined Network:\n")
gan.summary()
```

# 5. Training of GAN Model

## Configurations

- **`sample_images`**: Used to view the generated images and monitor the visual progress of the generator throughout training.
  
- **`save_checkpoint`**: Enables saving the model's state at each epoch, ensuring progress is not lost and allowing for continuity in training.
  
- **`load_checkpoint`**: Allows loading previously saved model weights, facilitating the resumption of training from the last saved state.


```python
def sample_images(noise, labels, subplots, figsize=(22, 8), save=False):
    generated_images = generator.predict([noise, labels])
    plt.figure(figsize=figsize)

    # Get discriminator output
    predicted_validity, predicted_labels = discriminator.predict([generated_images, labels])
    predicted_labels = np.argmax(predicted_labels, axis=1)
    predicted_validity = np.argmax(predicted_validity, axis=1)

    for i, (image,label) in enumerate(zip(generated_images, labels)):
        plt.subplot(subplots[0], subplots[1], i + 1)
        if CHANNELS == 1:
            plt.imshow(image.reshape((WIDTH, HEIGHT)), cmap='gray')
        else:
            plt.imshow(image.reshape((WIDTH, HEIGHT, CHANNELS)))

        title = f"Gen: {i}, Label: {label}\n"
        title += f"Pred Validity: {predicted_validity[i]}\n"
        title += f"Pred Label: {predicted_labels[i]}"
        
        plt.title(title)
        
        if save:
            img_name = f"gen_{i}_label_{label}.png"
            plt.savefig(img_name)
        plt.subplots_adjust(wspace=None, hspace=None)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Function to save checkpoints
def save_checkpoint(generator, discriminator, gan, epoch):
    checkpoint = {
        'generator': generator.get_weights(),
        'discriminator': discriminator.get_weights(),
        'gan': gan.get_weights(),
        'epoch': epoch
    }
    with open('gan_checkpoint_fine-tune.pkl', 'wb') as f:
        pickle.dump(checkpoint, f)

# Function to load checkpoints
def load_checkpoint(generator, discriminator, gan, filepath="gan_checkpoint_fine-tune.pkl"):
    if os.path.exists(filepath):
        with open(filepath, 'rb') as f:
            checkpoint = pickle.load(f)
            generator.set_weights(checkpoint['generator'])
            discriminator.set_weights(checkpoint['discriminator'])
            gan.set_weights(checkpoint['gan'])
            return checkpoint['epoch']
    return 0
```

## Training Phase

```python
from tqdm.auto import tqdm
np.random.seed(SEED)

# Load checkpoint if available
start_epoch = load_checkpoint(generator, discriminator, gan)

for epoch in range(start_epoch, EPOCHS):
    print(f"EPOCH {epoch + 1}/{EPOCHS}")
    
    for batch in tqdm(range(STEPS_PER_EPOCH)):
        try:
            # Generate random noise and random label for the batch size
            noise = np.random.normal(0, 1, (BATCH_SIZE, NOISE_DIM))
            random_labels = np.random.randint(0, NUM_CLASSES, BATCH_SIZE)
            
            # Generate fake images and the corresponding labels using the generator 
            fake_X = generator.predict([noise, random_labels])
            
            # Get a batch of real images and their labels
            real_X, real_labels = next(train_generator)

            # Create combined dataset of real and fake images
            X = np.concatenate((real_X, fake_X)) 
            y_real = np.ones((BATCH_SIZE, 1)) # Labels for real images
            y_fake = np.zeros((BATCH_SIZE, 1)) # Labels for fake images

            # Label Flipping
            if np.random.rand() < 0.05:
                y_real = np.zeros((BATCH_SIZE, 1)) # Labels for real images
                y_fake = np.ones((BATCH_SIZE, 1)) # Labels for fake images
                
            y_dis = np.concatenate((y_real, y_fake))
            labels = np.concatenate((real_labels, random_labels))
            
            # Unfreeze the discriminator weights
            discriminator.trainable = True

            # Train the discriminator
            d_loss = discriminator.train_on_batch([X, labels], [y_dis, labels])
            
            # Prepare labels for the generator
            y_gen = np.ones((BATCH_SIZE, 1))
            
            # Freeze the discriminator weights
            discriminator.trainable = False
            
            # Train the generator
            g_loss = gan.train_on_batch([noise, random_labels], [y_gen, random_labels])
        except Exception as e:
            print(f"Error during batch {batch + 1} in epoch {epoch + 1}: {e}")
            break
    
    print(f"EPOCH {epoch + 1}/{EPOCHS}")
    print(f"Generator Loss: Total: {g_loss[0]:.4f}, Adversarial: {g_loss[1]:.4f}, Class: {g_loss[2]:.4f}")
    print(f"Discriminator Loss: Total: {d_loss[0]:.4f}, Adversarial: {d_loss[1]:.4f}, Class: {d_loss[2]:.4f}")

    # Save checkpoint
    save_checkpoint(generator, discriminator, gan, epoch)
    
    # Generate sample images
    noise = np.random.normal(0, 1, size=(10, NOISE_DIM))
    labels = np.random.randint(0, NUM_CLASSES, 10)
    sample_images(noise, labels, (2, 5))
```

# 6. Evaluating the Discriminator

Only the discriminator is used, as the main focus was classification

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load saved model
load_checkpoint(generator, discriminator, gan)

# Collect all images and labels from the validation generator
def collect_all_validation_data(validation_generator):
    validation_images = []
    validation_labels = []
    
    # Loop over the validation generator to collect all images and labels
    for i in range(len(validation_generator)):
        images, labels = next(validation_generator)
        validation_images.append(images)
        validation_labels.append(labels)
    
    # Concatenate all batches
    validation_images = np.concatenate(validation_images)
    validation_labels = np.concatenate(validation_labels)
    
    return validation_images, validation_labels

# Collect all validation data
validation_images, validation_labels = collect_all_validation_data(validation_generator)

# Predict the validity and class labels using the discriminator
predicted_validity, predicted_labels = discriminator.predict([validation_images, validation_labels])

# Convert predicted_labels to the class with the highest probability
predicted_labels = np.argmax(predicted_labels, axis=1)

# Calculate evaluation metrics for each class
accuracy = accuracy_score(validation_labels, predicted_labels)
precision = precision_score(validation_labels, predicted_labels, average=None)
recall = recall_score(validation_labels, predicted_labels, average=None)
f1 = f1_score(validation_labels, predicted_labels, average=None)

# Print the metrics for each class
print(f'Accuracy: {accuracy:.4f}')
for i in range(NUM_CLASSES):
    print(f'Class {i} - Precision: {precision[i]:.4f}, Recall: {recall[i]:.4f}, F1 Score: {f1[i]:.4f}')

# Print Confusion Matrix
class_names = train_generator.class_indices
conf_matrix = confusion_matrix(validation_labels, predicted_labels)
print("Confusion Matrix:")
print(conf_matrix)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
